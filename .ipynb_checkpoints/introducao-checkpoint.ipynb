```python
#essa é uma celula que podemos rodar código. Aqui vamos importar os três módulos mais comuns para manipular dados.
import csv #esse ja vem com Python, os proximos três tem que ser instalados.
import numpy #trabalhar com vetores
import pandas #manipulacao de base da dados
import pylab #plotar graficos
```

Agora vamos fazer uma análise relativamente "simples" com dados do California PeMS. Trata-se de uma plataforma contendo dados de laços indutivos, magnetômetros e bluetooth mantido pela Caltrans. O California PeMS foi um dos primeiros sistemas deste tipo e é relevante até hoje pelo histórico e quantidade de sensores instalados e mantidos. Pesquisadores e profissionais da área podem fazer sua conta no California PeMS sem custo. Há um processo de liberação que no meu caso (alguns anos atrás, diga-se) foi rápido.

Os métodos, análise de limitações e tratamentos realizados com dados do PeMS é geral para dados de laços indutivos.

Os dados nos dois arquivos CSV's foram baixados no dia 22 de outubro de 2024 considerando o dia 21 de outubro de 2024.

```python
def ler_dados_como_csv_bruto(caminho_arquivo):
    #na pratica esse arquivo nao é um arquivo separado por virgula (comma separated value - csv), mas muito parecido. Tem um delimitador diferente, mas a ideia é parecida
    #a distribuicao típica do Python já vem com um modulo para abrir este tipo de arquivo numa série de tuplas ou dicionários - usaremos a forma de dicionário
    with open(caminho_arquivo, 'r') as f:
        reader = csv.DictReader(f, delimiter='\t')

        registros = []

        for linha in reader:
            registros.append(linha)
    
    return registros
```

```python
registros_jusante = ler_dados_como_csv_bruto('jusante.txt')
registros_montante = ler_dados_como_csv_bruto('montante.txt')

#vamos ver os primeiros 3 registros só para ver que deu certo
print(registros_jusante[0:3])
print(registros_montante[0:3])
```

```python

#podemos ler os horarios dos registros do texto da data (5-minutes), mas para simplificar vemos que temos registros a cada 5 minutos comecando 4 da manha
tempos = [4+i*5/60 for i in range(len(registros_jusante))] #colcando o tempo em horas

#vamos plotar um grafico agora
f, axs = pylab.subplots(1,2)

axs[0].plot(tempos, [float(rec['Occupancy (%)']) for rec in registros_jusante], label='Jusante')
axs[0].plot(tempos, [float(rec['Occupancy (%)']) for rec in registros_montante], label='Montante')
axs[1].plot(tempos, [(60/5)*float(rec['Flow (Veh/5 Minutes)']) for rec in registros_jusante], label='Jusante') #o 60/5=12 aqui vem da conversao do fluxo em veh/h
axs[1].plot(tempos, [(60/5)*float(rec['Flow (Veh/5 Minutes)']) for rec in registros_montante], label='Montante')
axs[0].legend()
axs[1].legend()
axs[0].set_xlabel("Horario (h)")
axs[1].set_xlabel("Horario (h)")
axs[0].set_ylabel("Ocupacao (%)")
axs[1].set_ylabel("Fluxo (veh/h)")
axs[0].grid(True)
axs[1].grid(True)
pylab.tight_layout()
pylab.show()
```

O código acima faz o básico de abrir um arquivo .csv (ou separado por tabulação) e plotar um gráfico. Uma coisa a se chamar atenção aqui é a conversão de ponto flutuante dos números (float(rec['Flow (Veh/5 Minutes)'])). Isto ocorre porque o módulo csv é simples e nem sempre trata bem campos numéricos e campos de texto. Há opções de diferentes 'quoting', mas por exemplo não deu certo para este caso porque os nomes dos campos não tem áspas.

Abaixo vamos repetir a mesma leitura usando uma biblioteca que trata melhor estas especifidades chamada pandas. É largamente utilizada na área de data science em Python.

```python
data_frame_jusante = pandas.read_csv("jusante.txt", delimiter="\t")
data_frame_montante = pandas.read_csv("montante.txt", delimiter="\t")

print(data_frame_jusante)
```

Perceba que ler os arquivos foi mais fácil e ele internamente já organizado em tabela e identificou corretamente os campos numéricos. Apesar de ter uma curva de aprendizado maior para algumas tarefas, essa biblioteca é uma ferramenta poderosa para tarefas comuns porém o "estilo" é um pouco diferente.

Utilizando uma lista de dicionário e fazendo operações mais "básicas", ainda que mais repetitivas, pode ser uma opção principalmente tratando de bases pequenas. No entanto, é bom enfatizar que a utilização de bibliotecas como o pandas para manipular dados não apresenta "desvantagens" (você pode fazer tudo através dela) e apresenta vantagens em vários casos. Portanto, vamos focar nela de agora em diante.

Vamos agora tentar tirar informações da rodovia através destes dados. Uma forma de calcular a capacidade de um trecho é observando o maior fluxo observado durante um período de 15 minutos. Como vemos que a jusante o fluxo é maior, vamos fazer isso nos dados a jusante da rampa de entrada.

```python
def media_movel(np_array, periodos):
    resultado = np_array.copy()
    for linha in range(periodos, np_array.shape[0]):
        resultado[linha] = numpy.sum(np_array[linha-periodos:linha])/periodos
    
    return resultado

lanes = 3
fluxos = (60/5)*data_frame_jusante["Flow (Veh/5 Minutes)"].to_numpy()
fluxos_media_movel = media_movel(fluxos, 3)

pylab.plot(tempos, fluxos, label='original')
pylab.plot(tempos, fluxos_media_movel, label='media movel')
pylab.legend()
pylab.xlabel("Horario (h)")
pylab.ylabel("Fluxo (veh/h)")
pylab.show()

print(f"O fluxo maximo da série original é {numpy.max(fluxos)} veic/h")
print(f"O fluxo maximo por faixa da série original é {numpy.max(fluxos)/lanes} veic/h.faixa")
print(f"O fluxo maximo da media movel é {numpy.max(fluxos_media_movel)} veic/h")
print(f"O fluxo maximo por faixa da media movel é {numpy.max(fluxos_media_movel)/lanes} veic/h.faixa")

```

```python
fluxo_montante = (1/3)*(60/5)*data_frame_montante["Flow (Veh/5 Minutes)"].to_numpy()
occ_montante = data_frame_montante["Occupancy (%)"].to_numpy()
fluxo_jusante = (1/3)*(60/5)*data_frame_jusante["Flow (Veh/5 Minutes)"].to_numpy()
occ_jusante = data_frame_jusante["Occupancy (%)"].to_numpy()

pylab.scatter(media_movel(occ_montante, 3), media_movel(fluxo_montante,3), label='Montante')
pylab.scatter(media_movel(occ_jusante, 3), media_movel(fluxo_jusante,3), label='Jusante')
pylab.ylim((0, 2000))
pylab.xlabel("Ocupação (%)")
pylab.ylabel("Fluxo (veh/h/faixa)")
pylab.legend()
pylab.show()
```

Nesse caso não faria tanta diferença, mas chegamos a conclusão que o máximo fluxo é em torno de 5040 veículos por hora o que dá em torno de 1680 veículos por faixa por hora. Um pouco abaixo do valor "típico" de 1800 veh/h/faixa e lembrando que esse valor típico talvez seja um pouco baixo.

Mas tem um pequeno detalhe digno de nota nos gráficos acima. Por que a parte não congestionada tem inclinações a montante e a jusante se as velocidades nesse regime devem ser muito próximas?

Aqui está um dos itens típicos de laço indutivos na medição de ocupação e seu significado. Quando medimos ocupação, o que tentamos inferir é a densidade de veículos. Para laços de comprimento D e se os veículos tem comprimento L, a relação entre ocupação e densidade se dá por:

Densidade = ocupação/(D+L).

A parte conveniente desta relação é que D+L é uma constante e portanto sempre podemos assumir que ocupação e densidade são grandezas proporcionais; o coeficiente angular é dado pela soma de comprimento médio dos veículos e do laço indutivo. No entanto há alguns cuidados que o analista deve tomar:

- O laço indutivo ainda que tenha um comprimento facilmente mensurável, a sua área de sensibilidade depende de múltiplas variáveis sendo duas que podem afetar os resultados. Primeiro, os laços indutivos tem configurações de sensibilidade e isto afeta a distância a distância limite entre o veículo para considerar presença de veículo.
- a quantidade e distribução de material magnético (aço) no veículo pode fazer com que dois veículos de mesmo comprimento tenham "comprimentos aparentes" diferentes, ou seja, para mesma velocidade os tempos de passagem podem ser distintos.

O provedor dos dados podem dar algumas informações. Neste caso específico esses comprimentos são dados como 0.3 e 0.5 no sistema do PeMS.

Se plotarmos os gráficos com o índice de correção, o resultado é esse:

```python
fluxo_montante = (1/3)*(60/5)*data_frame_montante["Flow (Veh/5 Minutes)"].to_numpy()
occ_montante = data_frame_montante["Occupancy (%)"].to_numpy()
fluxo_jusante = (1/3)*(60/5)*data_frame_jusante["Flow (Veh/5 Minutes)"].to_numpy()
occ_jusante = data_frame_jusante["Occupancy (%)"].to_numpy()

factor = 0.3/0.5
pylab.scatter(factor*media_movel(occ_montante, 3), media_movel(fluxo_montante,3), label='Montante')
pylab.scatter(media_movel(occ_jusante, 3), media_movel(fluxo_jusante,3), label='Jusante')
pylab.ylim((0, 2000))
pylab.xlabel("Ocupação (%)")
pylab.ylabel("Fluxo (veh/h/faixa)")
pylab.legend()
pylab.show()
```
